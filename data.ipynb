```python
import pandas as pd
import random
from datetime import datetime, timedelta
import pytz
```

```python
data = pd.read_csv('data/data.csv')
data
```

```python
def add_random_time_to_dates(df, date_col='date'):
    moscow_tz = pytz.timezone('Europe/Moscow')
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
    
    def random_timedelta():
        return timedelta(hours=random.randint(0, 23), minutes=random.randint(0, 59))
    
    df[date_col] = df[date_col].apply(lambda d: moscow_tz.localize(d + random_timedelta()) if not pd.isnull(d) else d)
    
    return df

# Assuming 'data' is your DataFrame and 'Date' is the name of your date column
data = add_random_time_to_dates(data, 'Date')
```

```python
data.drop(['Views'], axis=1, inplace=True)
```

```python
data.columns = ['title', 'date', 'location', 'department', 'amount']
# data['id'] = 0
data['encrypted_user_id'] = 0
data['description'] = '-'
new_column_order = ['encrypted_user_id', 'title', 'date', 'location', 'department', 'amount', 'description']
data = data[new_column_order]
data.head()
```
